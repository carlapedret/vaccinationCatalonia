- Upload csv
- Pivot table sum of the column by region and people vaccinated
- Add colum with population
- Calculate % vaccines over county population
- Eliminate columns we don't need
- Download the csv to create a map
   

```python
import pandas as pd
```

```python
# Upload the csv
'''The file is a csv and has some encoding problems. The last part of the code solves it.
Before, another solution was to upload it to Google Drive and downloaded again as a csv. 
It's separated by ; and it should be specified''' 

df = pd.read_csv('03_04_vacunes_diari.csv', sep=";", encoding = "ISO-8859-1", engine='python')
df.head(10)
```

```python
'''Delete the columns we don't need. In Pandas, rows (horizontal)
in a dataframe have axis=0, columns (vertical) have the first axis (axis=1).'''

df = df.drop(['DATA', 'SEXE', 'GRUP_EDAT', 'PCR', 'TAR', 
              'INGRESSOS_TOTAL','INGRESSOS_CRITIC', 'INGRESSATS_TOTAL'], 1) 
df.head(20)
```

```python
# We only need the columns Nom, vacunats i codi
df = df.drop(['CASOS_CONFIRMAT', 'INGRESSATS_CRITIC', 'EXITUS', 'CASOS_PCR', 'CASOS_TAR', 
              'POSITIVITAT_PCR_NUM','POSITIVITAT_TAR_NUM', 'POSITIVITAT_PCR_DEN', 'POSITIVITAT_TAR_DEN', 'CODI'], 1) 
df.head(20)
```

```python
# Create a new df based on a pivot table with the sum of the column 'Vacunats'
comarques = df.groupby('NOM')['VACUNATS_DOSI_1', 'VACUNATS_DOSI_2'].sum()
comarques.head(43)

#Another way to create a pivot table
#comarques = df.pivot_table(values="VACUNATS_DOSI_1", index="NOM", aggfunc='sum')
#comarques.head(17)
```

```python
#Import the second csv with the population data
df2 = pd.read_csv('poblacio.csv', sep=";")
df2.head(10)
```

```python
#Delete the columns you don't need
df2 = df2.drop(['Rt', 'TA Fets', 'Taxa confirmats per PCR/TA', 
              'Casos confirmats per PCR/TA','Taxa PCR/TA', 'PCR Fetes', '% PCR/TA Positives'], 1) 
df2.head(10)
```

```python
#Delete the rest of the columns
df2 = df2.drop(['Risc de rebrot', 'IncidÃ¨ncia acumulada a 14 dies', 'Vacunats', 'Ingressats', 'Defuncions'], 1) 
df2.head(10)
```

```python
#Merge the dataset with the vacunats and the datasets with the population
dfmerge = pd.merge(comarques, df2, left_on = 'NOM', right_on='Territori',validate='m:1', how='inner')
dfmerge.head(10)
```

```python
#Reorganize the columns
dfmerge = dfmerge[['Territori', 'PoblaciÃ³', 'VACUNATS_DOSI_1', 'VACUNATS_DOSI_2']]
dfmerge.head(10)
```

```python
#Calculate the national average
average = (dfmerge['VACUNATS_DOSI_1'].sum()/dfmerge['PoblaciÃ³'].sum())*100
print(average)
```

```python
#Check if the columns you want to make calculations with are intergers
dfmerge.dtypes
```

```python
#Calculate the percentage of population vaccinated
dfmerge['percentatge_poblaciÃ³_vacunada'] = (dfmerge['VACUNATS_DOSI_1']/dfmerge['PoblaciÃ³']) *100
dfmerge.head(10)
```

```python
# Reduce the number of decimal places
dfmerge['percentatge_poblaciÃ³_vacunada']=dfmerge['percentatge_poblaciÃ³_vacunada'].apply(lambda x:round(x,2))
dfmerge.head(10)
```
